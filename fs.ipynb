```python
import requests
import pandas as pd

# Set the API URL and your token
url = "https://fppdirectapi-prod.fuelpricesqld.com.au/Price/GetSitesPrices"
token = "5067f5e2-7e7f-49a7-bf4e-69b069e204c4"  # Replace with your actual token

# Set the headers for the request
headers = {
    "Authorization": f"FPDAPI SubscriberToken={token}",
    "Content-Type": "application/json"
}

# Add the query parameters
params = {
    "countryId": "21",
    "geoRegionLevel": "3",
    "geoRegionId": "1"
}

# Send the GET request to the API
response = requests.get(url, headers=headers, params=params)

# Check if the request was successful
if response.status_code == 200:
    fuel_data = response.json()
    print("Fuel price information retrieved successfully:")
    print(fuel_data)
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
    print(response.text)

# Convert the fuel_data into a DataFrame
df_fuel_prices = pd.DataFrame(fuel_data['SitePrices'])

# Display the DataFrame
print(df_fuel_prices)
```
